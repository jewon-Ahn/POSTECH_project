```python
# 'REPAIR_START_DATE'와 'REPAIR_END_DATE' 컬럼을 datetime 형식으로 변경
df1['REPAIR_START_DATE'] = pd.to_datetime(df1['REPAIR_START_DATE'], format='%Y%m%d')
df1['REPAIR_END_DATE'] = pd.to_datetime(df1['REPAIR_END_DATE'], format='%Y%m%d')

# 'MONTH_YYYYMM' 컬럼을 'mm' 형식으로 변경하고 int로 변환
df1['MONTH_YYYYMM'] = df1['MONTH_YYYYMM'].astype(str).str[4:].astype(int)

# 'REPAIR_PERIOD' 컬럼 생성 및 값 할당
df1['REPAIR_PERIOD'] = df1['REPAIR_END_DATE'] - df1['REPAIR_START_DATE']
```

```python
# 빈도가 상위 10% 이상인 관측치들만 추출

# MODEL_CODE 컬럼의 빈도 확인
model_code_counts = df1['MODEL_CODE'].value_counts()

# 빈도가 상위 10%에 해당하는 값을 가져옵니다.
top_10_percent_model_codes = model_code_counts[model_code_counts >= model_code_counts.quantile(0.99)].index.tolist()

# 상위 10%에 해당하는 관측치만 추출
df_top_10_percent = df1[df1['MODEL_CODE'].isin(top_10_percent_model_codes)]
df_top_10_percent
```
