```python
import pandas as pd
import numpy as np
import pyarrow as pa
import pyarrow.parquet as pq
import matplotlib.pyplot as plt
```

```python
parquet_file = 'C:/Users/USER/OneDrive - postech.ac.kr/바탕 화면/LG/postech_company_project.parquet'
# Parquet 파일 열기
table = pq.read_table(parquet_file)

# Table에서 데이터를 DataFrame으로 변환
df = table.to_pandas()
df.head()
```

```python
df['DIVISION_NAME']
```

```python
df['DIVISION_NAME'].value_counts()
```

```python
# 'DIVISION_NAME', 'PRODUCT_NAME', 'PRODUCT_NAME_DETAIL'을 기준으로 데이터를 그룹화합니다.
grouped_data = df.groupby(['DIVISION_NAME', 'PRODUCT_NAME', 'PRODUCT_NAME_DETAIL'])

# 각 그룹에 속하는 데이터의 개수를 확인합니다.
grouped_counts = grouped_data.size().reset_index(name='Count')
grouped_counts=grouped_counts[::-1]
grouped_counts= grouped_counts.reset_index(drop=True)

# 결과를 출력합니다.
grouped_counts
```
