```python
import pandas as pd
import numpy as np
import re
from tqdm import tqdm
```

```python
t = pd.read_parquet('postech_company_project.parquet')
```

```python
t['RECEIPT_NUMBER_IDX'] = t['RECEIPT_NUMBER'].str.replace('LGE_REC_', '').astype(int, errors='ignore')
t['PREV_RECEIPT_NUMBER_IDX'] = t['PREV_RECEIPT_NUMBER'].str.replace('LGE_REC_', '').astype(int, errors='ignore')

t.set_index('RECEIPT_NUMBER_IDX', inplace=True)
```

```python
t1 = t[t['RECLAIM'] == 1]

# for rows in t1
# trace back the previous receipt number and update the TO_RECLAIM column to 1
# and add receipt number to the row
t['TO_RECLAIM'] = 0
t['TO_RECLAIM_NUMBER'] = np.nan

for index, row in tqdm(t1.iterrows(), total = t1.shape[0]):
    prev_receipt = int(row['PREV_RECEIPT_NUMBER_IDX'])
    if prev_receipt in t.index:
        t.at[prev_receipt, 'TO_RECLAIM'] = 1
        t.at[prev_receipt, 'TO_RECLAIM_NUMBER'] = index
```

```python
# save dataframe as parquet file
t.to_parquet('postech_company_project_2.parquet')
```
